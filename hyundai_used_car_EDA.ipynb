# Four Scatter plots with numeric Columns

```python

plt.figure(figsize=(20,16))

plt.subplot(4, 1, 1)
plt.title("Fig:  Prices of cars with respect to Mileage")
plt.subplots_adjust(hspace=0.4)
sns.scatterplot(x="mileage", y="price", data=df)
plt.xlabel("Mileage")
plt.ylabel("Price")

plt.subplot(4, 1, 2)

plt.title("Fig:  Prices of cars with respect to  MPG")

sns.scatterplot(x="mpg", y="price", data=df)
plt.xlabel("Mile Per Galon")
plt.ylabel("Price")
plt.subplot(4, 1, 3)
plt.title("Fig:  Prices of cars with respect to  Engine Size")

sns.scatterplot(x="engineSize", y="price", data=df)
plt.xlabel("Engine Size")
plt.ylabel("Price")

plt.subplot(4, 1, 4)
plt.title("Fig:  Prices of cars with respect to FuelType")

sns.scatterplot(x="fuelType", y="price", data=df)
plt.xlabel("Fuel Type")
plt.ylabel("Price")

plt.show()


```

Scatter plot is a graph of two sets of data along the two axes. It is used to visualize the relationship between the two variables. If the value along the Y axis seem to increase as X axis increases(or decreases), it could indicate a positive (or negative) linear relationship.


# Two Regression plot with numeric cloumns

```python
plt.figure(figsize=(20,16))

plt.subplot(2, 1, 1)
plt.title("Fig:  Prices of cars with respect to Mileage")

sns.regplot(x="mileage", y="price", data=df,line_kws={"color":"red"})
plt.xlabel("Mileage")
plt.ylabel("Price")

plt.subplot(2,1,2)
plt.title("Fig:  Prices of cars with respect to Engine Size")

sns.regplot(x="engineSize",y="price",data=df,line_kws={"color":"red"})
plt.xlabel("Engine Size")
plt.ylabel("Price")

plt.show()
```

The above figures  are examples of two scatter plots and the same plots with a linear regression line added. The regression line is an attempt to find the best fit through the points in the scatter plot. We may probably able to discern a linear relationship from the first plot but adding a regression line on top of the scatter diagram makes it much clearer.



# Pair Plot with Numeric Columns

```python
sns.pairplot(df,hue="fuelType")
plt.show()
```

The pairplot plot is shown in the image above. Its using the hyundai data set. The data set has 5 measurements: Year, Price,mileage, MPG, and enginesize respectively.

This type of analysis is  used in machine learning  because the measurements  provide an excellent way to established relationship of numeric values with each others. The data is mapped in the grid . Because there are 5 measurements, it creates a 5X5 plot.
