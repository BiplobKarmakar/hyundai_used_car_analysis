#(A) Load the data


```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from google.colab import files
 
 
uploaded = files.upload()
```

# A1. Check Data in the dataset

```python
df=pd.read_csv("hyundai.csv")
display(df.tail())
```

# A2. Columns and its type checking for choosing features

```python
df.info()
```

# (B) Separating the numeric features and target variable

```python
features = ['year', 'mileage','mpg', 'engineSize']
target = ['price']

X = df[features]
y = df[target]

print("Shape of X = ", X.shape)
print("Shape of y = ", y.shape)
```

# (C) Split the original dataset into the train set (80%) and the test set (20%) 


```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    train_size=0.8, 
                                                    random_state=42)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# (D) Performing Linear Regression and Predict the 'Price' from the test set,

```python
model = LinearRegression()
model = model.fit(X_train, y_train)
```

y = m1x1 + m2x2 + m3x3 + m4x4 + c and all four values of m



```python
coefficients = model.coef_
print(coefficients)
```

Value of c

```python
c = model.intercept_
print(c)
```

```python
X.columns
```

```python
coef_df = pd.DataFrame({"features": X.columns, 
                       "coefficients": np.squeeze(coefficients)})

display(coef_df)
```

A positive sign indicates that as the feature variable increases, the target variable also increases.

A negative sign indicates that as the feature variable increases, the target variable decreases.

# D 1. Predicting the 'Price' from the test set

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
print(y_test)
```

#(E) Finding  the RMSE, MAE, MAPE, MSE, and R squared KPIs from the actual test data and the predicted data

```python
MAE = mean_absolute_error(y_test, y_pred)
print("MAE = ", MAE)

MSE = mean_squared_error(y_test, y_pred, squared=True)
print("MSE = ", MSE)

RMSE = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE = ", RMSE)

r2 = r2_score(y_test, y_pred)
print("r_squared = ", r2) 

MAPE=np.mean(np.abs((y_test - y_pred) / y_test)) * 100
#MAPE=np.mean(np.abs((y_pred - y_test) / y_pred)) * 100

print("MAPE = ", MAPE)
```

MAE : Mean Absolute Error (MAE) is calculated by taking the summation of the absolute difference between the actual and calculated values of each observation over the entire array and then dividing the sum obtained by the number of observations in the array.

 MSE :the mean squared error or mean squared deviation of an estimator measures the average of the squares of the errorsthat is, the average squared difference between the estimated values and the actual value. MSE is a risk function, corresponding to the expected value of the squared error loss.

  RMSE: Root Mean Square Error (RMSE) is the standard deviation of the residuals (prediction errors). Residuals are a measure of how far from the regression line data points are; RMSE is a measure of how spread out these residuals are.

 r_squared =  R-squared measures the strength of the relationship between your linear model and the dependent variables on a 0 - 100% scale

 MAPE: Mean Absolute Percentage Error (MAPE) is a statistical measure to define the accuracy of a machine learning algorithm on a particular dataset.
